### Pyplot

The `pyplot` module in Matplotlib provides a simple interface for creating plots and visualizations. Let's import it.

````python
# Import the pyplot module and alias it as plt.
import matplotlib.pyplot as plt

years=list(map(str,range(1980,2014)))
df.loc["Afghanistan",years].plot(kind='line')
df.loc['Egypt',years].plot(kind='line')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.title('Afghanistan vs Egypt Immigrants')
plt.legend(['Afghanistan','Egypt'])
plt.show()
````

### Example: Drawing a Line
Use `plt.plot()` to draw a line between two points.

````python
# Draw a line in a diagram from position (0,0) to position (6,250):

# Import NumPy for numerical operations.
import numpy as np

# Define x-coordinates of the line.
xpoints = np.array([0, 6])
# Define y-coordinates of the line.
ypoints = np.array([0, 250])

# Plot the line using xpoints and ypoints.
plt.plot(xpoints, ypoints)  
# Display the plot.
plt.show()
````

### Plotting Without a Line

You can visualize points without connecting them using a line.

````python
# Plotting Without Line
# Define x-coordinates for points.
xpoints = np.array([1, 8])
# Define y-coordinates for points.
ypoints = np.array([3, 10])

# Use 'o' to plot only the points.
plt.plot(xpoints, ypoints, 'o')
# Display the plot.
plt.show()
````

### Multiple Points

Plot multiple points and connect them with a line.

````python
# Multiple Points
# You can plot as many points as you like, just make sure you have the same number of points in both axis.

# Define x-coordinates for multiple points.
xpoints = np.array([1, 2, 6, 8])

# Define corresponding y-coordinates.
ypoints = np.array([3, 8, 1, 10])

# Plot the points and connect them with a line.
plt.plot(xpoints, ypoints)

# Display the plot.
plt.show()
````

### Default X-Points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, (etc. depending on the length of the y-points.

````python
# Define y-coordinates for the points.
ypoints = np.array([3, 8, 1, 10, 5, 7])

# Plot the points with default x-coordinates.
plt.plot(ypoints)   #default: 0 1 2 3 4 5

# Display the plot.
plt.show()
````

## Markers
Markers are used to emphasize points in a plot.

````python
# You can use the keyword argument marker to emphasize each point with a specified marker:

# Define y-coordinates.
ypoints = np.array([3, 8, 1, 10])

# Add circle markers to each point.
plt.plot(ypoints, marker = 'o')

# Display the plot.
plt.show()
````

````python
# Use a star as a marker.
plt.plot(ypoints, marker = '*')
````

### Formatting Markers and Lines

Customize markers, lines, and colors using the `fmt` parameter.

````python
# Format Strings fmt
# This parameter is also called fmt, and is written with this syntax:
# marker|line|color
ypoints = np.array([3, 8, 1, 10])

# Circle markers, dashed line, and red color.
plt.plot(ypoints, 'o:r')

# Display the plot.
plt.show()
````

### Marker Size

Adjust the size of the markers using the `ms` (marker size) parameter.

````python
# Set marker size to 20: marker size is set using the keyword argument ms.
plt.plot(ypoints, marker = 'o', ms = 20)

# Display the plot.
plt.show()
````

### Marker Color

Customize marker edge color using `mec` and face color using `mfc`.

````python
#Marker Color
# Red edge.
# marker edge color is set using the keyword argument mec.
plt.plot(ypoints, marker = 'o', mec = 'r')

# Display the plot.
plt.show()
````

````python
# You can use the keyword argument markerfacecolor or the shorter mfc 
# to set the color inside the edge of the markers:
# Red face.
plt.plot(ypoints, marker = 'o', mfc = 'r')

# Display the plot.
plt.show()
````

````python
# Set the color of both the edge and the face to red:
# Red edge and face.
plt.plot(ypoints, marker = 'o', mec = 'r', mfc = 'r')

# Display the plot.
plt.show()
````

## Line Properties

Adjust the thickness of the line using `linewidth`.

````python
# Set line width to 2.5.
plt.plot(ypoints, linewidth = 2.5)

# Display the plot.
plt.show()
````

## Multiple Lines

You can plot multiple lines in a single chart.

````python
# Multiple Lines

# First line data.
y1 = np.array([3, 8, 1, 10])

# Second line data.
y2 = np.array([6, 2, 7, 11])

# Plot the first line.
plt.plot(y1)

# Plot the second line.
plt.plot(y2)

# Display the plot.
plt.show()
````

````python
# Alternatively, you can explicitly define both `x` and `y` for each line.

# x-coordinates for the first line.
x1 = np.array([0, 1, 2, 3])

# y-coordinates for the first line.
y1 = np.array([3, 8, 1, 10])

# x-coordinates for the second line.
x2 = np.array([0, 1, 2, 3])

# y-coordinates for the second line.
y2 = np.array([6, 2, 7, 11])

# Plot both lines on the same chart.
plt.plot(x1, y1, x2, y2)

# Display the plot.
plt.show()
````

## Adding Titles and Labels to Plots

Titles and labels help make your plots more descriptive and informative.

````python
# Create a title and add labels to the axes.

# X-axis data points.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Y-axis data points.
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Plot the line using x and y data points.
plt.plot(x, y)

# Add a title to the plot.
plt.title('Title of the Plot')

# Label the x-axis.
plt.xlabel('x-axis')

# Label the y-axis.
plt.ylabel('y_axis')

# Display the plot.
plt.show()
````

````python
# Set font properties for the title and labels:

# X-axis data points.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Y-axis data points.
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Set font properties for the title and labels.

# Define font properties for the title.
font1={'family':'serif','color':'purple','size':20}

# Define font properties for labels.
font2={'family':'serif','color':'gray','size':20}


# Apply font1 to the title.
plt.title('Title of the Plot', fontdict = font1)

# Apply font2 to the x-axis label.
plt.xlabel('x-axis', fontdict = font2)

# Apply font2 to the y-axis label.
plt.ylabel('y_axis', fontdict = font2)

# Plot the line using x and y data points.
plt.plot(x, y)

# Display the plot.
plt.show()
````

````python
# Position the title at the left of the plot.

# X-axis data points.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Y-axis data points.
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Position the title at the left of the plot.
plt.title('Title of the Plot', loc = 'left')

# Add x-axis label.
plt.xlabel('x-axis')

# Add y-axis label.
plt.ylabel('y_axis')

# Plot the line.
plt.plot(x, y)

# Display the plot.
plt.show()
````

## Adding Grid Lines

Grid lines help to make the plot easier to read by highlighting the graph's scale.

````python
# Add grid lines to the plot.

# X-axis data points.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

# Y-axis data points.
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Add a title.
plt.title('Title of the Plot')

# Label the x-axis.
plt.xlabel('x-axis')

# Label the y-axis.
plt.ylabel('y_axis')

# Plot the data points.
plt.plot(x, y)

# Add grid lines.
plt.grid()

# Display the plot.
plt.show()
````

## Scatter Plots

Scatter plots show the relationship between two sets of data points.

````python
# Create a scatter plot.
# X-axis data points.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

# Y-axis data points.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Create the scatter plot.
plt.scatter(x, y)

# Display the plot.
plt.show()
````

````python
# Draw two scatter plots on the same figure with different colors:
# X-axis data points.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

# Y-axis data points.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Scatter plot for the first dataset with the color hotpink.
plt.scatter(x, y, color = 'hotpink')

# X-axis data points.
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])

# Y-axis data points.
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

# Scatter plot for the second dataset with the color code #88c999.
plt.scatter(x, y, color = '#88c999')

# Display the plot.
plt.show()
````

````python
# Adding a color map to scatter plots
# X-axis data points.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

# Y-axis data points.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Assign a color value for each point.
colors=np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# Use the "viridis" colormap for the points.
plt.scatter(x, y, c = colors, cmap = 'viridis')

# Add a color scale bar for reference.
plt.colorbar()

# Display the plot.
plt.show()
````

````python
# Adding a color map to scatter plots
# X-axis data points.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

# Y-axis data points.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Assign a color value for each point.
plt.scatter(x, y, c = y)

# Use the "viridis" colormap for the points.
plt.colormap = 'viridis'

# Add a color scale bar for reference.
plt.colorbar()

# Display the plot.
plt.show()
````

````python
# Set custom marker sizes for the scatter plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

# Y-axis data points.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Assign marker sizes.
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

# Use the "s" argument to set marker sizes.
plt.scatter(x, y, s = sizes)

# Display the plot.
plt.show()
````

## Bars

Visualize categorical data with bar charts.

````python
# Categories on the x-axis.
x = np.array(["A", "B", "C", "D"])

# Values for each category.
y = np.array([3, 8, 1, 10])

# Create the bar chart.
plt.bar(x, y)

# Display the chart.
plt.show()
````

````python
# Customize the color of the bars.
# Set the bar color to red.
plt.bar(x, y, color = 'red')

# Display the chart.
plt.show()
````

````python
# Adjust the width of the bars.
# Categories on the x-axis.
x = np.array(["A", "B", "C", "D"])

# Values for each category.
y = np.array([3, 8, 1, 10])

# Reduce the width of the bars.
plt.bar(x, y, color='green', width = 0.2)

# Display the chart.
plt.show()
````

## Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

Shows the data distribution

````python
# Generate a dataset with a normal distribution.
x = np.random.normal(170, 10, 250)

# Print the generated data points.
print(x)
````

````python
# Create a histogram to visualize the distribution of the data.
# Plot the histogram.
#no need for 'y' it is the freq it calculates automatically -->used in univariate analysis
plt.hist(x)

# Display the plot.
plt.show()
````

Example: Say you ask for the height of 250 people, you might end up with a histogram like this:

* 2 people from 140 to 145cm
* 5 people from 145 to 150cm
* 15 people from 151 to 156cm
* 31 people from 157 to 162cm

## Pie Charts

Pie charts represent parts of a whole.

````python
# Create a simple pie chart.

# Values for each pie slice.
y = np.array([35, 25, 25, 15])

# Create the pie chart.
plt.pie(y)

# Display the chart.
plt.show()
````

````python
# Add labels to the pie chart.
# Values for each slice.
y = np.array([35, 25, 25, 15])

# Labels for each slice.
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Create the pie chart with labels.
plt.pie(y, labels = mylabels)

# Display the chart.
plt.show()
````

````python
# Add labels to the pie chart with explode property
# Values for each slice.
y = np.array([35, 25, 25, 15])

# Labels for each slice.
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
exp_list = [0.2, 0.1, 0.1, 0.1]
# Create the pie chart with labels.
plt.pie(y, labels = mylabels, explode=exp_list, shadow=True)

# Display the chart.
plt.show()
````
